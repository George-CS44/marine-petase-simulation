```python
#{ run: "auto" }
# Drag these sliders to test different ocean conditions
temperature_celsius = 8 #{type:"slider", min:0, max:40, step:1}
ocean_ph = 8.1 #{type:"slider", min:6.0, max:9.0, step:0.1}

import numpy as np
import matplotlib.pyplot as plt
from src.temperature_model import activity_vs_temperature
from src.ph_model import activity_vs_ph

wild_type_params = {
    'name': 'Wild PETase Type',
    'Ea': 65000,
    'T_opt': 35,
    'T_den': 50,
    'Vmax': 0.8,
    'pH_opt': 7.5,
    'pH_sigma': 1.5,
    'Km': 0.8,
    'colour': '#2196F3'
}

marine_mutant_params = {
    'name': 'Marine PETase Type',
    'Ea': 4200,
    'T_opt': 12,
    'T_den': 25,
    'Vmax': 60,
    'pH_opt': 8.1,
    'pH_sigma': 0.6,
    'Km': 0.25,
    'colour': '#4CAF50'
}

wt_temp_factor = activity_vs_temperature(temperature_celsius, wild_type_params) / wild_type_params['Vmax']
wt_ph_factor = activity_vs_ph(ocean_ph, wild_type_params) / wild_type_params['Vmax']
wt_combined_pct = (wt_temp_factor * wt_ph_factor) * 100

mut_temp_factor = activity_vs_temperature(temperature_celsius, marine_mutant_params) / marine_mutant_params['Vmax']
mut_ph_factor = activity_vs_ph(ocean_ph, marine_mutant_params) / marine_mutant_params['Vmax']
mut_combined_pct = (mut_temp_factor * mut_ph_factor) * 100

fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(
    [wild_type_params['name'], marine_mutant_params['name']], 
    [wt_combined_pct, mut_combined_pct], 
    color=[wild_type_params['colour'], marine_mutant_params['colour']], 
    edgecolor='black', 
    width=0.5
)

ax.set_ylabel('Relative Activity (% of Maximum)')
ax.set_title(f'Enzyme Activity Comparison at {temperature_celsius}°C, pH {ocean_ph}')
ax.set_ylim(0, 110)

for bar in bars:
    yval = bar.get_height()
    text_y = min(yval + 2, 105)
    ax.text(bar.get_x() + bar.get_width()/2.0, text_y, f'{yval:.1f}%', ha='center', va='bottom', fontweight='bold')

plt.show()

print(f"Wild-Type Final Percentage: {wt_combined_pct:.2f}%")
print(f"Marine Mutant Final Percentage: {mut_combined_pct:.2f}%")
```
